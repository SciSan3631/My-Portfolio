```python
import os
import pandas as pd
```

```python
df = pd.read_csv('labels_mbrset.csv')
```

```python
df.columns
```

```python
import plotly.express as px

# Assuming 'final_artifacts' is the column with artifact categories
fig = px.pie(df, names='final_artifacts', title='Distribution of Final Artifacts')
fig.show()
```

```python
fig = px.pie(df, names='final_quality', title='Distribution of Final Quality')
fig.show()
```

```python
cross_table = pd.crosstab(df['final_quality'], df['final_artifacts'])
print(cross_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'final_quality' and 'final_artifacts' are your column names
cross_table = pd.crosstab(df['final_quality'], df['final_artifacts'])

# Create a heatmap
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.heatmap(cross_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Cross-Tabulation of Final Quality and Final Artifacts")
plt.xlabel("Final Artifacts")
plt.ylabel("Final Quality")
plt.show()
```
